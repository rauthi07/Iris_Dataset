```python
df.plot(kind='scatter',x='sl',y='sw')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='species',size=6).map(plt.scatter,'sl','sw').add_legend()
plt.show()
```

```python
sns.pairplot(df,hue='species',size=3).add_legend()
plt.show()
```

```python
iris_setosa=df.loc[df["species"]=="setosa"];
iris_virginica=df.loc[df["species"]=="virginica"];
iris_versicolor=df.loc[df["species"]=="versicolor"];
```

```python
plt.plot(iris_setosa['pl'],np.zeros_like(iris_setosa['pl']),'o')
plt.plot(iris_virginica['pl'],np.zeros_like(iris_virginica['pl']),'o')
plt.plot(iris_versicolor['pl'],np.zeros_like(iris_versicolor['pl']),'o')
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'pl').add_legend()
plt.show()
sns.distplot(iris_setosa['pl'])
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'pw').add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'sw').add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'sl').add_legend();
plt.show()
```

```python
count,bin_edges=np.histogram(iris_setosa['pl'],bins=10,density=True)
pdf=count/(sum(count))
plt.plot(bin_edges[1:],pdf);
plt.show()
```

```python
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf);
plt.show()
```
